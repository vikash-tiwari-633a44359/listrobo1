```python
class Car:
    def __init__(self, name, model, color):
        self.name = name
        self.model = model
        self.color = color

    def get_info(self):
        return f"{self.name} {self.model} {self.color}"


car1 = Car("Toyota", "Camry", "Red")
print(car1.get_info())
```
